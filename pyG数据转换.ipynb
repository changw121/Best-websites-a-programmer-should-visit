* pre_transform函数
对整个数据集做预处理，比如：对所有的节点进行一个固定的特征映射，或者对所有的图进行归一化。只会在读取数据集时调用一次，所以常用来做结点特征的归一化。
pre_transform接受一个Data对象作为参数，该对象表示一个包含结点和边特征的图。函数可以对这个Data对象进行添加、删除、修改结点和边的特征的操作。最后，数据集中的每个图都会被影响。

* transform函数
它可以对每个图进行不同的转换，因为它会在每次数据集被调用时进行，它可以：为节点特征添加噪声或对边特征进行采样。
trasform的参数与pre_transform一样，能进行的操作也一样，不同的是，它只会影响当前调用的图，而不是整个数据集。
